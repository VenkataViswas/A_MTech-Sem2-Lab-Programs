# Remove Special Characters from text


```python
def remove_special_characters(text):
    ans=""
    for char in text:
      if char.isalnum() or char.isspace():
        ans+=char
    return ans


input_text = "this$ is an exa@mple Text ^for testing% !!! "
output_text = remove_special_characters(input_text)
print(output_text)
```

## Implementing a regular expression to parse all the plural words


```python
import re
word = "foods"
result = re.search(r"s$",word)
if result:
  print("its a plural ")
else :
  print("its not plural")
```
