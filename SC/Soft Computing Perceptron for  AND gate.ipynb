Importing Required libraries


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Training the perceptron model

```python
def train_perceptron(X, y, learning_rate=0.1, epochs=20):
    n_samples, n_features = X.shape
    weights = np.zeros(n_features)
    bias = 0.0
    errors_per_epoch = []

    for _ in range(epochs):
        errors = 0
        for xi, target in zip(X, y):
            linear_output = np.dot(xi, weights) + bias
            y_pred = 1 if linear_output >= 0 else 0
            update = learning_rate * (target - y_pred)
            weights += update * xi
            bias += update
            errors += int(update != 0)
        errors_per_epoch.append(errors)

    return weights, bias, errors_per_epoch
```

```python
def predict_perceptron(X, weights, bias):
    linear_output = np.dot(X, weights) + bias
    return np.where(linear_output >= 0, 1, 0)
```

```python
X_and = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y_and = np.array([0, 0, 0, 1])

weights, bias, errors = train_perceptron(X_and, y_and, learning_rate=0.1, epochs=20)

predictions = predict_perceptron(X_and, weights, bias)

print("Errors in each epochs :",errors)
print("after all the epochs final Weights:", weights)
print("after all the epochs final Bias:", bias)
print("after all the epochs final Predictions:", predictions)
```
