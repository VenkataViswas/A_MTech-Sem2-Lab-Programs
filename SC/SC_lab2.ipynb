## Fuzzy Sets


```python
def fuzzy_union(A, B):
    union_set = {}
    for element in A.keys():
        union_set[element] = max(A[element], B[element])
    return union_set
```

```python
def fuzzy_intersection(A, B):
    intersection_set = {}
    for element in A.keys():
        intersection_set[element] = min(A[element], B[element])
    return intersection_set
```

```python
def fuzzy_complement(A):
    complement_set = {}
    for element, membership in A.items():
        complement_set[element] = 1.0 - membership
    return complement_set
```

```python

def fuzzy_difference(A, B):
    difference_set = {}
    complement_B = fuzzy_complement(B)
    return fuzzy_intersection(A,complement_B)
```

```python
A = {
    'x1': 0.8,
    'x2': 0.3,
    'x3': 0.7
}

B = {
    'x1': 0.6,
    'x2': 0.9,
    'x3': 0.2
}

print("Fuzzy Set A:", A)
print("Fuzzy Set B:", B)

```

```python
print("\n-- Operations --")

#  Union
union_AB = fuzzy_union(A, B)
print(" Union (A U B):", {k: round(v, 3) for k, v in union_AB.items()})

#  Intersection
intersection_AB = fuzzy_intersection(A, B)
print(" Intersection (A intersect B):", {k: round(v, 3) for k, v in intersection_AB.items()})

# 3) Complement (A')
complement_A = fuzzy_complement(A)
print(" Complement of A (A'):", {k: round(v, 3) for k, v in complement_A.items()})

# 4) Set Difference (A - B)
difference_AB = fuzzy_difference(A, B)
print(" Set Difference (A - B):", {k: round(v, 3) for k, v in difference_AB.items()})
```
